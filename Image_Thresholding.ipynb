```
"""from google.colab import drive
drive.mount('/content/gdrive') """
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

drive_image = "/content/gdrive/MyDrive/IMG_9393.JPG" 
#drive_image = get_ipython().getoutput(![](https://drive.google.com/open?id=1wIc8OfYaHokNn6vCeX_hUkNJmZky6ZJr))
image = cv.imread(drive_image)
plt.imshow(image)
plt.title("ReadImage : BGR")
plt.show()



image_RGB = cv.cvtColor(image,cv.COLOR_BGR2RGB)
plt.imshow(image_RGB)
plt.title("Converted to RGB")
plt.show()

image_gs = cv.imread(drive_image,0)
plt.imshow(image_gs)
plt.title("ImageRead : BGR,Grayscale")
plt.show()

koala = 10
no_koala = 0
for x in range(koala):
  no_koala=no_koala+1
print("nigga \
 Koala : " , no_koala , "koala again")

image_gray = cv.imread(drive_image,0)
image_RGB2 = cv.cvtColor(image_gray,cv.COLOR_BGR2RGB)
Title = "Image read : BGR,Graysacle : Converted to Grayscale"
plt.imshow(image_RGB2)
plt.title(Title)
plt.show()


Threshold_Value = 127
max_value = 255

ret,o1 = cv.threshold(image_RGB2,Threshold_Value,max_value,cv.THRESH_BINARY)
ret,o2 = cv.threshold(image_RGB2,Threshold_Value,max_value,cv.THRESH_BINARY_INV)
ret,o3 = cv.threshold(image_RGB2,Threshold_Value,max_value,cv.THRESH_TOZERO)
ret,o4 = cv.threshold(image_RGB2,Threshold_Value,max_value,cv.THRESH_TOZERO_INV)
ret,o5 = cv.threshold(image_RGB2,Threshold_Value,max_value,cv.THRESH_TRUNC)
Title = ["Binary","Inverted_Binary","TOZERO","TOZERO_INVERTED","TRUNC"]
output = [o1,o2,o3,o4,o5]

print("Is it ?")
for x in range(5):
  
  plt.imshow(output[x])
  plt.title(Title[x])
  plt.show()
```
